```python
# Task 1: Exploring and Visualizing a Simple Dataset (Iris Dataset)

# Step 0: Install libs if needed (uncomment if running in a fresh environment)
# !pip install pandas seaborn matplotlib

# Step 1: Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# For inline plotting in Jupyter/Colab
%matplotlib inline

# Make plots readable
plt.rcParams.update({'figure.max_open_warning': 0})

# Step 2: Load dataset (seaborn has built-in iris)
df = sns.load_dataset("iris")

# Optional: save a CSV copy
df.to_csv("iris_dataset.csv", index=False)

# Step 3: Basic inspection
print("âœ… Dataset loaded")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("\nFirst 5 rows:")
display(df.head())

# Step 4: Info and describe
print("\nInfo:")
df.info()
print("\nDescriptive statistics:")
display(df.describe())

# Extra: species-wise means
print("\nMean values by species:")
display(df.groupby('species').mean())

# Step 5: Scatter plot (sepal length vs sepal width)
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species", s=70)
plt.title("Scatter: Sepal Length vs Sepal Width (by species)")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid(alpha=0.3, linestyle='--')
plt.show()

# Short insight:
print("Insight: Setosa usually has smaller sepal length; there is partial separation by species.")

# Step 6: Pairplot (pairwise relationships)
sns.pairplot(df, hue="species", plot_kws={'s':30})
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()
print("Insight: Petal length and petal width give clear separation between species.")

# Step 7: Histograms (distributions)
numeric_cols = ['sepal_length','sepal_width','petal_length','petal_width']
df[numeric_cols].hist(figsize=(10,8), bins=12, edgecolor='black')
plt.suptitle("Histograms of numerical features")
plt.show()
print("Insight: Look for skewness/multi-modality in distributions.")

# Step 8: Box plots (outliers & spread)
plt.figure(figsize=(10,6))
sns.boxplot(data=df[numeric_cols])
plt.title("Box plots (features)")
plt.show()
print("Insight: Boxplots show spread and potential outliers across features.")

# Step 9: Species-wise boxplots (compare species)
plt.figure(figsize=(12,8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2,2,i)
    sns.boxplot(x='species', y=col, data=df)
    plt.title(f"{col} by species")
    plt.xlabel("")
plt.tight_layout()
plt.show()
print("Insight: Compare medians and spread across species; petal features vary more between species.")

# Final message
print("\nâœ… EDA completed. Save notebook as 'iris_eda.ipynb' and upload to GitHub under Task_1_Iris_EDA/")
```
